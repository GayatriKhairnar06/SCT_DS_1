```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Provisional_COVID-19_Deaths_by_Sex_and_Age.csv")
```

```python
print(df.head)
```

```python
#bar chart
#keep only required columns
df1 = df[["Sex","Age Group", "COVID-19 Deaths"]]
grouped  = df1.groupby(["Age Group","Sex"])["COVID-19 Deaths"].sum().unstack()
#plotting
grouped.plot(kind="bar",figsize=(10,6), edgecolor="red")

plt.title("COVID-19 Deaths by Sex and Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Deaths")
plt.xticks(rotation=45)
plt.legend(title="Sex")
plt.tight_layout()
plt.show()
```

```python
# Select the 3 numeric columns
cols = ["COVID-19 Deaths", "Pneumonia Deaths", "Influenza Deaths"]

# Create subplots (1 row, 3 columns)
plt.figure(figsize=(18, 5))

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    plt.hist(df[col].dropna(), bins=10, color="skyblue", edgecolor="black")
    plt.title(f"Distribution of {col}")
    plt.xlabel("Number of Deaths")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
